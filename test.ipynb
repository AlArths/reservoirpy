```python
import matplotlib.pyplot as plt
import numpy as np

from reservoirpy.nodes import Reservoir
```

```python
node1 = Reservoir(100, lr=0.8, input_bias=True, seed=1234)
node2 = Reservoir(50, lr=0.5, input_bias=True, seed=1234)

node1 = node1 << node2

data = np.sin(np.linspace(1, 100, 1000)).reshape(-1, 1)
res = (node1 >> node2).run(data)
```

```python
_ = plt.plot(data)
```
